```python
metal = ['Sc', 'Ti', 'V','Cr','Mn','Fe','Co','Ni','Cu','Zn', \
         'Y','Zr','Nb','Mo','Ru','Rh','Pd','Ag','Cd','Hf',\
         'Ta','W','Re','Os','Ir','Pt','Au','La','Ce']   # 29
# 1-10

nonmetal = ['O'] 
# 1-10

second = ['Cr','Mn','Fe','Co','Ni']

element_ratio = [0.1, 0.2, 0.5]
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

result = [(1, 1)]
for i in range(1, 11):
    for j in range(i+1, 11):
        if gcd(i, j) == 1:
            result.append((i, j))  # 32
```

```python
compound_O = []
for nm in nonmetal:
    for m in metal:
        for index in result:
            if index[0]==1 and index[1]==1:
                com = m + nm
            elif index[0] == 1:
                com = m + nm + f'{index[1]}'
            elif index[1] == 1:
                com = m + f'{index[0]}' + nm
            else:
                com = m + f'{index[0]}' + nm + f'{index[1]}'
            compound_O.append(com)
# compound
# compound = pd.DataFrame(compound, columns=['compound'])
# compound.to_csv('NEW_TMCs.csv', index=False)
compound_O
```

```python
import math

Bicompound_O = []

for nm in nonmetal:
    for m in metal:
        for se in second:
            if se == m:
                continue
            for index in result:
                for er in element_ratio:
                    se_index = round((index[0] * er), 1) if (index[0] * er) != 1 else ''
                    m_index = round((index[0]*(1-er)), 1) if (index[0]*(1-er)) != 1 else '' 
                    nm_index = index[1] if index[1] != 1 else ''
                    com = se + str(se_index) + m + str(m_index) + nm + str(nm_index)
                    com_same = m + str(se_index) + se + str(m_index) + nm + str(nm_index)
                    if com_same not in Bicompound_O:
                        Bicompound_O.append(com)
#Bicompound_O = pd.DataFrame(Bicompound_O, columns=['compound'])
# compound.to_csv('NEW_TMCs.csv', index=False)
Bicompound_O
```

```python
import re

original_combinations = ["M1N1", "M2N1", "M3N2", "M1S2"]

compound_CNS = []  
index_CNS = []  

for original_combination in original_combinations:
    for replacement in metal:
        modified_combination = original_combination.replace("M", replacement, 1)

        matches = re.findall(r'(\d+)', modified_combination)

        if matches:
            indices = tuple(map(int, matches)) 

        index_CNS.append(indices)
        modified_combination_without_ones = re.sub(r'1(?!\d)', '', modified_combination)
        compound_CNS.append(modified_combination_without_ones)
```

```python
import re
import json

def parse_compound(compound):
    matches = re.findall(r'([A-Z][a-z]*)(\d*\.*\d*)', compound)
    
    result = {}
    
    for match in matches:
        element, count = match
        count = float(count) if count else 1.0 
        result[element] = count
    
    return result

def get_res(compound_name):
    parsed_result = parse_compound(compound_name)
    return parsed_result
```

```python

Bicompound_CNS = []
for ratio in element_ratio:
    for compound in compound_CNS:
        com = get_res(compound)
        key = list(com.keys())
        val = list(com.values())
        for za in second:
            if key[0] != za:
                za_index = round(ratio * val[0], 1)
                single_index = round((1-ratio) * val[0], 1)
                tail_index = int(val[1])
                if float(za_index) == 1.0:
                    za_index = ''
                if float(single_index) == 1.0:
                    single_index = ''
                if float(tail_index) == 1.0:
                    tail_index = ''
                new_com = za + str(za_index) + key[0] + str(single_index) + key[1] + str(tail_index)
                Bicompound_CNS.append(new_com)
Bicompound_CNS
```
